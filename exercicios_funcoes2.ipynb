 Crie uma função com parÂmetro e sem retorno chamada mostrar_tabuada(numero) que recebe um parâmetro inteiro e imprime na tela a tabuada de 1 a 10 desse número.

```python
def mostrar_tabuada(numero):
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")

mostrar_tabuada(8)
```

3) Crie uma função com retorno e com parâmetro chamada validar_senha(senha) que recebe uma string e retorna um dicionário contendo:

Chave Significado

tamanho_valido    Verdadeiro se tiver ≥ 8 caracteres
tem_numero  Verdadeiro se possuir número
tem_maiuscula     Verdadeiro se tiver letra maiúscula
tem_minuscula     Verdadeiro se tiver letra minúscula
tem_especial      Verdadeiro se possuir caractere especial
senha_valida      Verdadeiro apenas se todos os critérios forem atendidos

```python
def validar_senha(senha):
    resultado = {
        "tamanho_valido": len(senha) >= 8,
        "tem_numero": False,
        "tem_maiuscula": False,
        "tem_minuscula": False,
        "tem_especial": False, }

    for c in senha:
        if c.isdigit():
            resultado["tem_numero"] = True
        elif c.isupper():
            resultado["tem_maiuscula"] = True
        elif c.islower():
            resultado["tem_minuscula"] = True
        elif not c.isalnum():
            resultado["tem_especial"] = True

    resultado["senha_valida"] = all(resultado.values())

    return resultado
```
